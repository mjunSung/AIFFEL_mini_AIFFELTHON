```python
import pandas as pd
import re

# 파일 경로 리스트 (필요한 파일 경로 5개를 입력하세요)
file_paths = [
    '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/AI 산업 기술.txt',
    '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/AI 영상인식 음성인식.txt',
    '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/AI 윤리.txt',
    '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/AI 의료.txt',
    '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/AI 자율주행.txt'
]

# 각 파일을 처리하고 CSV로 저장
for i, file_path in enumerate(file_paths):
    # 파일 읽기
    with open(file_path, 'r', encoding='utf-8') as file:
        lines = file.read()

    # 기사 구분 및 DataFrame 재구성
    data = []
    # 기사 블록 나누기 (빈 줄 기준으로 나눔)
    articles = re.split(r'\n\n+', lines)

    for article in articles:
        # 제목, URL, 본문 추출 (제목: 첫 줄, URL: 두 번째 줄, 나머지는 본문)
        parts = article.strip().split('\n', 2)
        if len(parts) == 3:
            title, url, content = parts
            data.append({'Title': title, 'URL': url, 'Content': content})
        elif len(parts) == 2:
            title, url = parts
            data.append({'Title': title, 'URL': url, 'Content': ''})

    # DataFrame 생성
    df = pd.DataFrame(data)

    # CSV 파일로 저장
    output_path = f'/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/processed_article_{i + 1}.csv'
    df.to_csv(output_path, index=False, encoding='utf-8-sig')
    print(f"Processed and saved file {i + 1} to {output_path}")
```

```python
import pandas as pd
import re

# 파일 경로
file_path = '/Users/mj/Library/CloudStorage/OneDrive-개인/AI_study/crawling/all_articles.txt'

# 파일 읽기
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.read()

# 기사 구분 및 DataFrame 재구성
data = []

# 기사 블록 나누기 (빈 줄 기준으로 나눔)
articles = re.split(r'\n\n+', lines)

for article in articles:
    # 제목, URL, 본문 추출 (제목: 첫 줄, URL: 두 번째 줄, 나머지는 본문)
    parts = article.strip().split('\n', 2)
    if len(parts) == 3:
        title, url, content = parts
        data.append({'Title': title, 'URL': url, 'Content': content})
    elif len(parts) == 2:
        title, url = parts
        data.append({'Title': title, 'URL': url, 'Content': ''})

# DataFrame 생성
df = pd.DataFrame(data)

# 결과 확인
df.head()
```

```python
import pandas as pd
from konlpy.tag import Okt
from rouge import Rouge
from nltk.translate.bleu_score import sentence_bleu
from nltk.translate.meteor_score import meteor_score
from bert_score import score

okt = Okt()

```

```python
nltk.download('wordnet')
```

```python
# 1. 형태소 기반 텍스트 토큰화
def tokenize_korean(text):
    return okt.morphs(text)

# 2. ROUGE
rouge = Rouge()
df['rouge_scores'] = df.apply(
    lambda row: rouge.get_scores(row['generated_answer'], row['answer'])[0],
    axis=1
)

# 3. BLEU
df['bleu_score'] = df.apply(
    lambda row: sentence_bleu(
        [tokenize_korean(row['answer'])], 
        tokenize_korean(row['generated_answer'])
    ),
    axis=1
)

# 4. METEOR
df['meteor_score'] = df.apply(
    lambda row: meteor_score(
        [tokenize_korean(row['answer'])], 
        tokenize_korean(row['generated_answer'])
    ),
    axis=1
)

# 5. BERTScore
P, R, F1 = score(
    df['generated_answer'].tolist(), 
    df['answer'].tolist(), 
    lang="ko",  # 한글 데이터
    verbose=True
)
df['bertscore_precision'] = P.tolist()
df['bertscore_recall'] = R.tolist()
df['bertscore_f1'] = F1.tolist()
```
